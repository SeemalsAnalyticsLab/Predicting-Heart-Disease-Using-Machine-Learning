This project will demonstrate how to predict heart disease using a machine learning classification model.

We'll go through the whole process, from data analysis to model building, and explain each step in an easy-to-understand way with visualizations.

# **🚀 Overview**
**Problem Definition:** Can we predict whether a patient has heart disease based on clinical features?

**Data:** We'll work with a dataset containing 14 clinical attributes.

**Evaluation:** We'll use accuracy as our metric and aim for a high-performance model.

**Tools:** We'll use Python libraries such as pandas, NumPy, Matplotlib, and Scikit-Learn.

# **👣 Steps**
   Here are all the steps:

- **Step 1: Problem Definition**
- **Step 2: Data Collection**
- **Step 3: Data Exploration**
- **Step 4: Data Preprocessing**
- **Step 5: Model Training**
- **Step 6: Model Improvement**
- **Step 7: Feature Importance Visualization**

### **🧠 Step 1: Understanding the Data**
First, let's load and explore the data.

```python
# Importing library
import pandas as pd

# Loading the dataset
df = pd.read_csv("heart dataset.csv")

# Display the shape of the DataFrame (number of rows and columns)
df.shape  # (rows, columns)
```

Next, let's preview the first few rows.

```python
# Showing the first 10 rows of the DataFrame
df.head(10)
```

### **🧐 Data Summary:**

- **Age:** 👴👵 Age of the patient.
  
- **Sex:** 🚹 1 = Male, 0 = Female.
  
- **Cp (Chest Pain Type):** 💔 Ranges from 0-3, indicating different types of chest pain.
  
- **Trestbps:** 🩺 Resting blood pressure.
  
- **Chol:** 🥑 Cholesterol level.
  
- **Target:** 🏥 1 = Heart disease, 0 = No heart disease.

### **📊 Step 2: Visualizing Target Distribution**
Let’s visualize how many people in our dataset have heart disease versus those who don’t.

```python
# Ploting a bar chart to visualize the count of target classes
df['target'].value_counts().plot(kind='bar', color=['teal', 'coral'])

```

This bar plot helps us quickly understand how balanced our dataset is.

In this case, more people have heart disease than those without.

### **🔍 Step 3: Exploratory Data Analysis (EDA)**
We'll now dive deeper into the dataset and look at how features like age, sex, and cholesterol correlate with heart disease.

**Age Distribution** 👴👵

```python
# Ploting a histogram to show the distribution of patients' ages
df['age'].plot(kind='hist', color='teal')

```

This histogram shows the distribution of patients' ages in the dataset.

**Heart Disease💓 vs. Age**👴👵

```python
# Importing library
import matplotlib.pyplot as plt

# Creating a scatter plot to visualize the relationship between age, cholesterol, and heart disease
plt.scatter(df['age'], df['chol'], c=df['target'], cmap='Set1')
plt.xlabel('Age')  # Label for x-axis
plt.ylabel('Cholesterol')  # Label for y-axis
plt.title('Age vs. Cholesterol with Heart Disease')  # Title of the plot
```

This scatter plot shows the relationship between age, cholesterol, and heart disease.

Color indicates whether a patient has heart disease.

###**🤖 Step 4: Machine Learning Model**
Now, let's train a simple machine learning model using Scikit-Learn. We'll start with a logistic regression model.

```python
# Importing libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Spliting the dataset into features (X) and target (y)
X = df.drop('target', axis=1)  # Features (independent variables)
y = df['target']  # Target (dependent variable)

# Spliting the data into training and testing sets (80% training, 20% testing)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Creating and train the logistic regression model
model = LogisticRegression(max_iter=1000)  # Increase max_iter for convergence
model.fit(X_train, y_train)  # Fit the model on the training data

# Making predictions on the test set
y_pred = model.predict(X_test)

# Calculating accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")  # Printing accuracy as a percentage
```

**✅ Model Evaluation**

We achieve an accuracy score, which tells us how well our model is performing.


### **🛠️ Step 5: Improving the Model**
We'll try different machine learning models, such as Decision Trees, Random Forest, or Gradient Boosting, to see if we can improve the performance.

```python
# Importing library
from sklearn.ensemble import RandomForestClassifier

# Creating and train the Random Forest model
model = RandomForestClassifier()
model.fit(X_train, y_train)  # Fiting the model on the training data
```

### **📉 Step 6: Feature Importance**
Let’s see which features are most important for predicting heart disease.


```python
# Geting feature importances from the trained Random Forest model
importances = model.feature_importances_
feature_names = X.columns  # Geting the names of the features

# Creating a sorted list of features based on their importance
sorted(zip(importances, feature_names), reverse=True)
```

### **Step 7: Visualizing Feature Importances 📊🔍**
Let’s explore which features play a crucial role in predicting heart disease and understand their significance in our model.

```python
#Importing library
import matplotlib.pyplot as plt
import pandas as pd

# Geting feature importances from the trained Random Forest model
importances = model.feature_importances_
feature_names = X.columns  # Get the names of the features

# Creating a sorted list of features based on their importance
sorted_importances = sorted(zip(importances, feature_names), reverse=True)

# Unziping the sorted list into two lists: features and their corresponding importances
importances_sorted, feature_names_sorted = zip(*sorted_importances)

# Creating a DataFrame for better visualization
feature_importances_df = pd.DataFrame({
    'Feature': feature_names_sorted,
    'Importance': importances_sorted
})

# Ploting the feature importances
plt.figure(figsize=(10, 6))
plt.barh(feature_importances_df['Feature'], feature_importances_df['Importance'], color='teal')
plt.xlabel('Importance')
plt.title('Feature Importances from Random Forest Classifier')
plt.gca().invert_yaxis()  # Inverting y-axis to show the most important feature at the top
plt.show()
```

This code visualizes the importance of features in predicting heart disease using a horizontal bar chart from a trained Random Forest model.

Cholesterol and chest pain type seem to have strong predictive power.
